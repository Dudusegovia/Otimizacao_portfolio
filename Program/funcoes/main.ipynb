```python
import numpy as np

def func(X):
    # Calculo da função objetivo
    nmax = 50
    npsel = len(X)  # Assuming npsel is the length of X
    E = np.zeros(100)
    Cov = np.zeros((100, 100))
    Retorno = 0.0

    # Retorno
    for i in range(npsel):
        Retorno += X[i] * E[i]

    # Funcional das covariancias:
    Q = 0.0
    for i in range(npsel):
        for j in range(npsel):
            Q += X[i] * Cov[i, j] * X[j]

    # Restrição de percentagens (soma 1): 
    g = 0.0
    for i in range(npsel):
        g += X[i]

    # Pseudo-função objetivo penalizada:
    rp = 1.0  # Assuming a value for rp, as it is not defined in the original code
    OBJ = 10.0 - Retorno + Q + rp * abs(g)

    # Acrescente restrição de positividade das variáveis
    for i in range(npsel):
        OBJ += rp * abs(min(0.0, X[i]))  # Restrições de positividade das variáveis

    return OBJ
```
